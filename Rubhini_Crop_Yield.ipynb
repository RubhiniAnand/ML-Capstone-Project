# 5.REMOVING THE OUTLIERS

** Function is created to get columns separately based on the dtypes **

```python
def quanQual(Crop_raw_ds):
    quan=[]
    qual=[]
    for columnname in Crop_raw_ds.columns:
            #print(columnname)
        if(Crop_raw_ds[columnname].dtype=='O'):
            #print("qual")
            qual.append(columnname)
        else:
           # print("quan")
            quan.append(columnname)
    return quan,qual
        
```

```python
quan,qual=quanQual(Crop_raw_ds)
```

```python
quan
```

```python
qual
```

```python
Crop_raw_ds.describe()
```

```python
descriptive=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","99%","Q4:100%","IQR","1.5rule","Lesser","Greater","Min","Max"],columns=quan)
for columnName in quan:
    descriptive.loc["Mean",columnName]=Crop_raw_ds[columnName].mean()
    descriptive.loc["Median",columnName]=Crop_raw_ds[columnName].median()
    descriptive.loc["Mode",columnName]=Crop_raw_ds[columnName].mode()[0]
    descriptive.loc["Q1:25%",columnName]=Crop_raw_ds.describe()[columnName]["25%"]
    descriptive.loc["Q2:50%",columnName]=Crop_raw_ds.describe()[columnName]["50%"]
    descriptive.loc["Q3:75%",columnName]=Crop_raw_ds.describe()[columnName]["75%"]
    descriptive[columnName]["99%"]=np.percentile(Crop_raw_ds[columnName],99)
    descriptive.loc["Q4:100%",columnName]=Crop_raw_ds.describe()[columnName]["max"]   
    descriptive.loc["IQR",columnName]=descriptive[columnName]["Q3:75%"]-descriptive[columnName]["Q1:25%"]
    descriptive.loc["1.5rule",columnName]=1.5*descriptive[columnName]["IQR"]
    descriptive.loc["Lesser",columnName]=descriptive[columnName]["Q1:25%"]-descriptive[columnName]["1.5rule"]
    descriptive.loc["Greater",columnName]=descriptive[columnName]["Q3:75%"]+descriptive[columnName]["1.5rule"]
    descriptive.loc["Min",columnName]=Crop_raw_ds[columnName].min()
    descriptive.loc["Max",columnName]=Crop_raw_ds[columnName].max()

   
```

```python
descriptive
```

```python
lesser=[]
greater=[]
for columnName in quan:
    if(descriptive[columnName]["Min"]<descriptive[columnName]["Lesser"]):
        lesser.append(columnName)
    if(descriptive[columnName]["Max"]>descriptive[columnName]["Greater"]):
        greater.append(columnName)
        
```

```python
lesser
```

```python
greater
```

```python
for column in lesser:
    Crop_raw_ds[column][Crop_raw_ds[column]<descriptive[column]["Lesser"]]=descriptive[column]["Lesser"]
for column in greater:
    Crop_raw_ds[column][Crop_raw_ds[column]>descriptive[column]["Greater"]]=descriptive[column]["Greater"]
```

```python
descriptive=pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%","Q3:75%","99%","Q4:100%","IQR","1.5rule","Lesser","Greater","Min","Max"],columns=quan)
for columnName in quan:
    descriptive.loc["Mean",columnName]=Crop_raw_ds[columnName].mean()
    descriptive.loc["Median",columnName]=Crop_raw_ds[columnName].median()
    descriptive.loc["Mode",columnName]=Crop_raw_ds[columnName].mode()[0]
    descriptive.loc["Q1:25%",columnName]=Crop_raw_ds.describe()[columnName]["25%"]
    descriptive.loc["Q2:50%",columnName]=Crop_raw_ds.describe()[columnName]["50%"]
    descriptive.loc["Q3:75%",columnName]=Crop_raw_ds.describe()[columnName]["75%"]
    descriptive[columnName]["99%"]=np.percentile(Crop_raw_ds[columnName],99)
    descriptive.loc["Q4:100%",columnName]=Crop_raw_ds.describe()[columnName]["max"]   
    descriptive.loc["IQR",columnName]=descriptive[columnName]["Q3:75%"]-descriptive[columnName]["Q1:25%"]
    descriptive.loc["1.5rule",columnName]=1.5*descriptive[columnName]["IQR"]
    descriptive.loc["Lesser",columnName]=descriptive[columnName]["Q1:25%"]-descriptive[columnName]["1.5rule"]
    descriptive.loc["Greater",columnName]=descriptive[columnName]["Q3:75%"]+descriptive[columnName]["1.5rule"]
    descriptive.loc["Min",columnName]=Crop_raw_ds[columnName].min()
    descriptive.loc["Max",columnName]=Crop_raw_ds[columnName].max()

   
```

```python
descriptive
```

```python
lesser=[]
greater=[]
for columnName in quan:
    if(descriptive[columnName]["Lesser"]>descriptive[columnName]["Min"]):
        lesser.append(columnName)
    if(descriptive[columnName]["Greater"]<descriptive[columnName]["Q4:100%"]):
        greater.append(columnName)   
```

# 6.ONE HOT ENCODING TECHNIQUE

**Categorical Data are of two types:**
1.Nominal(One Hot Encoder)  2.Ordinal(Label Encoder)

One hot encoding is a technique that we use to represent categorical 
variables as numerical values in a machine learning model.
